```python
def mostrar_tablero(tablero):
    print("╔═══╦═══╦═══╗")  # borde superior
    for fila, i in enumerate(tablero): # fila es cada file (0,1,2), i es cada indice dentro de cada fila
        print("║ " + " ║ ".join(i) + " ║")
        if fila < len(tablero) - 1:
            print("╠═══╬═══╬═══╣")  
    print("╚═══╩═══╩═══╝") # borde inferior 
        
mostrar_tablero([[' ', ' ', ' '], [' ', ' ', ' '], [' ', ' ',' ']])
```

```python

#diccionario con simbolos
simbolos = {'X': 'X', 'O': '●', ' ': ' '}

def verificar_ganador(tablero):
    #verifica filas
    for fila in tablero:
        if fila[0] == fila[1] == fila[2] != simbolos[' ']: #3 elementos de la fila son iguales? ignorando vacios
            return fila[0] #retorna el simbolo del ganador
    
    # Verifica columnas
    for columna in range(3):
        if tablero[0][columna] == tablero[1][columna] == tablero[2][columna] != simbolos[' ']:
            return tablero[0][columna]

    # Verifica diagonales
    if tablero[0][0] == tablero[1][1] == tablero[2][2] != simbolos[' ']:
        return tablero[0][0]
    if tablero[0][2] == tablero[1][1] == tablero[2][0] != simbolos[' ']:
        return tablero[0][2]

    # Si nadie gana
    return None
    
    
```

```python
import tkinter as tk
from tkinter import messagebox

class TresEnRaya:
    def __init__(self, root):
        self.root = root
        self.root.title("Tres en raya")
        self.current_player = "X"
        self.board = [None] * 9
        self.buttons = []
        self.score = {"X": 0, "O": 0, "Empates": 0}

        self.build_ui()
        self.update_status()

    def build_ui(self):
        # Frame tablero
        board_frame = tk.Frame(self.root)
        board_frame.pack(padx=10, pady=10)

        for i in range(9):
            btn = tk.Button(board_frame, text="", font=("Helvetica", 30),
                            width=3, height=1,
                            command=lambda i=i: self.on_click(i))
            btn.grid(row=i//3, column=i%3, padx=5, pady=5)
            self.buttons.append(btn)

        # Panel inferior: estado, reiniciar y marcador
        bottom = tk.Frame(self.root)
        bottom.pack(padx=10, pady=(0,10), fill="x")

        self.status_label = tk.Label(bottom, text="", font=("Helvetica", 14))
        self.status_label.pack(side="left")

        restart_btn = tk.Button(bottom, text="Reiniciar partida", command=self.reset_board)
        restart_btn.pack(side="right", padx=5)

        reset_score_btn = tk.Button(bottom, text="Reiniciar marcador", command=self.reset_score)
        reset_score_btn.pack(side="right", padx=5)

        # Marcador arriba
        score_frame = tk.Frame(self.root)
        score_frame.pack(padx=10, pady=(5,0))
        self.score_label = tk.Label(score_frame, text=self.get_score_text(), font=("Helvetica", 12))
        self.score_label.pack()

    def get_score_text(self):
        return f"Marcador — X: {self.score['X']}    O: {self.score['O']}    Empates: {self.score['Empates']}"

    def update_status(self):
        self.status_label.config(text=f"Turno de: {self.current_player}")
        self.score_label.config(text=self.get_score_text())

    def on_click(self, index):
        if self.board[index] is not None:
            return  # ya marcado

        self.board[index] = self.current_player
        self.buttons[index].config(text=self.current_player, state="disabled")

        winner = self.check_winner()
        if winner:
            self.handle_win(winner)
        elif all(cell is not None for cell in self.board):
            self.handle_draw()
        else:
            self.current_player = "O" if self.current_player == "X" else "X"
            self.update_status()

    def check_winner(self):
        b = self.board
        wins = [
            (0,1,2), (3,4,5), (6,7,8),  # filas
            (0,3,6), (1,4,7), (2,5,8),  # columnas
            (0,4,8), (2,4,6)            # diagonales
        ]
        for (i,j,k) in wins:
            if b[i] and b[i] == b[j] == b[k]:
                return b[i]
        return None

    def handle_win(self, player):
        self.score[player] += 1
        self.update_status()
        messagebox.showinfo("¡Ganador!", f"¡{player} ha ganado!")
        self.after_round_reset()

    def handle_draw(self):
        self.score["Empates"] += 1
        self.update_status()
        messagebox.showinfo("Empate", "¡La partida ha terminado en empate!")
        self.after_round_reset()

    def after_round_reset(self):
        # deshabilitar botones temporalmente y reiniciar tablero
        for b in self.buttons:
            b.config(state="disabled")
        # reiniciamos tras pequeña pausa para que el usuario vea resultado
        self.root.after(500, self.reset_board)

    def reset_board(self):
        self.board = [None] * 9
        for btn in self.buttons:
            btn.config(text="", state="normal")
        # empieza el siguiente turno con el jugador que perdió la última o alternamos:
        # aquí dejamos al jugador X iniciar siempre; si quieres alternar, modifica esto.
        self.current_player = "X"
        self.update_status()

    def reset_score(self):
        if messagebox.askyesno("Confirmar", "¿Reiniciar marcador?"):
            self.score = {"X": 0, "O": 0, "Empates": 0}
            self.update_status()

if __name__ == "__main__":
    root = tk.Tk()
    app = TresEnRaya(root)
    root.mainloop()
```
